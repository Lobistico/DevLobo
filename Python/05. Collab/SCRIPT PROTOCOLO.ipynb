```python
import pandas as pd
import sqlite3
```

```python
conexao = sqlite3.connect('basededados.db')
cursor = conexao.cursor()

cursor.execute('CREATE TABLE IF NOT EXISTS planilha ('
                'id INTEGER PRIMARY KEY AUTOINCREMENT,'
                'processo TEXT,'
                'assunto TEXT,'
                'data text,'
                'tipo text'
                # 'a_tempo TEXT'
                ')')
teste ={} # Salva um dicionário teste com temperatura e estação.
predominio=[] # Salva o predominio do momento.
sair = True
while sair == True:
    processo = input('Qual processo?')
    assunto = input('Qual assunto?')
    data = input('Qual é a data?')
    tipo = input('TIPO')

    cursor.execute('INSERT INTO planilha (processo, assunto, data, tipo) VALUES (?, ?, ?,?)', (processo,assunto,data,tipo))
    saida = input('sair?')
    if saida == 'Y':
      sair = False
    # print(f' Na estação {dc_cid_cod[root[i][0].text]} está fazendo {root[i][3].text} graus')  -- > É um teste para verificar se está funcionando o código
conexao.commit()
# PARA VER COMO ESTÁ SENDO SALVO 
# cursor.execute('SELECT * FROM previsao_tempo')
# for linha in cursor.fetchall():
#     print(linha)
cursor.close()
conexao.close()
```

```python
#RECUPERAÇÃO DE DADOS - SALVA EM CSV
conect = sqlite3.connect('/content/basededados.db')

dados_recuperados = pd.read_sql_query("SELECT * FROM planilha", conect)
dados_recuperados.to_csv('dados.csv',index=False)
```
